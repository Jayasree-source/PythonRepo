### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows in the dataset: {duplicate_count}")

# Display the first few duplicate rows
duplicates = df[df.duplicated(keep=False)]
print("First few duplicate rows:")
print(duplicates.head())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
df_cleaned = df.drop_duplicates()

# Verify removal by counting duplicates again
duplicate_count_after = df_cleaned.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicate_count_after}")

# Optional: Display the first few rows of the cleaned dataset
print("Preview of cleaned dataset:")
print(df_cleaned.head())
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
missing_remote = df['RemoteWork'].isnull().sum()
print(f"Number of missing rows in 'RemoteWork': {missing_remote}")
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
df['RemoteWork'].value_counts()
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
most_frequent_value = df['RemoteWork'].mode()[0]  # mode()[0] gives the top value
print(f"The most frequent value in 'RemoteWork' is: {most_frequent_value}")
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
majority_value = df['RemoteWork'].mode()[0]

# Impute missing values with the majority value
df['RemoteWork'].fillna(majority_value, inplace=True)

# Verify that missing values are handled
missing_after = df['RemoteWork'].isnull().sum()
print(f"Missing values in 'RemoteWork' after imputation: {missing_after}")
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
## Write your code here

import matplotlib.pyplot as plt
import seaborn as sns

# Identify compensation-related columns
comp_cols = [col for col in df.columns if 'Comp' in col or 'Salary' in col]
print("Compensation-related columns found:", comp_cols)
comp_cols = ['CompTotal', 'ConvertedCompYearly']

#compensation-related columns description
df['CompTotal'].describe()
df['ConvertedCompYearly'].describe()
# Visualize each compensation column
for col in comp_cols:
    plt.figure(figsize=(12, 5))
        
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col].dropna(), kde=True, bins=30, color='skyblue')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col].dropna(), color='lightgreen')
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    
    plt.tight_layout()
    plt.show()
```
