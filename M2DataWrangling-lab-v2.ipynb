### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

```python
# Write your code here
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.notnull().sum(),
    'Missing Values': df.isnull().sum()
})
print(summary)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# Example for 'Country' column
print(df['Country'].value_counts())  # Check unique values and frequency

# Example for 'EdLevel' column
print(df['EdLevel'].value_counts())

df['Country'].value_counts().to_csv('country_value_counts.csv', header=['Count'])

# Export EdLevel value counts to CSV
df['EdLevel'].value_counts().to_csv('edlevel_value_counts.csv', header=['Count'])
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# Example mapping for Country
country_mapping = {
   "United States of America": "USA",
    "United Kingdom of Great Britain and Northern Ireland": "UK",
    "Russian Federation": "Russia",
    "Iran, Islamic Republic of...": "Iran",
    "Venezuela, Bolivarian Republic of...": "Venezuela",
    "CÃ´te d'Ivoire": "Cote d'Ivoire",
    "Democratic Republic of the Congo": "DR Congo",
    "Republic of Korea": "South Korea",
    "North Korea": "North Korea",
    "Hong Kong (S.A.R.)": "Hong Kong",
    "Taiwan": "Taiwan",
    "Nomadic": "Other"
}

df['Country'] = df['Country'].map(lambda x: country_mapping.get(x, x))  # map values, keep unchanged if not in dictionary

# Example mapping for Education Level
edlevel_mapping = {
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Bachelor's",
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Master's",
    "Some college/university study without earning a degree": "Some college",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "High School",
    "Professional degree (JD, MD, Ph.D, Ed.D, etc.)": "Professional",
    "Associate degree (A.A., A.S., etc.)": "Associate",
    "Primary/elementary school": "Primary",
    "Something else": "Other"
}

df['EdLevel'] = df['EdLevel'].map(lambda x: edlevel_mapping.get(x, x))
print(df['Country'].value_counts())
print(df['EdLevel'].value_counts())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
# One-hot encode the Employment column
employment_dummies = pd.get_dummies(df['Employment'], prefix='Employment')
#print(employment_dummies)
# Concatenate the one-hot encoded columns back to the original dataframe
df = pd.concat([df, employment_dummies], axis=1)

# Check the result
print(df.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# Count missing values in each column
missing_counts = df.isnull().sum()

# Sort columns by number of missing values in descending order
missing_counts_sorted = missing_counts.sort_values(ascending=False)

# Display the columns with the most missing values
print(missing_counts_sorted.head(10))  # top 10 columns with most missing values
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# Impute missing values with mean
df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean(), inplace=True)

# Alternatively, impute missing values with median
# df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median(), inplace=True)

# Verify missing values are filled
print(df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# Find the most frequent value (mode) in the column
most_frequent = df['RemoteWork'].mode()[0]

# Impute missing values with the most frequent value
df['RemoteWork'].fillna(most_frequent, inplace=True)

# Verify missing values are filled
print(df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
!pip install scikit-learn
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Fill missing values first
df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median(), inplace=True)

# Min-Max scaling manually
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / \
                                   (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])

# Check the results
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
#df['YearsCodePro'].value_counts().to_csv('Years_Code_Pro.csv', header=['Count'])
# First, ensure 'YearsCodePro' is numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Define ExperienceLevel based on YearsCodePro
def categorize_experience(years):
    if pd.isna(years):
        return 'Unknown'
    elif years < 2:
        return 'Junior'
    elif 2 <= years < 5:
        return 'Mid'
    elif 5 <= years < 10:
        return 'Senior'
    else:
        return 'Expert'

# Apply the function to create the new column
df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)

# Check the result
df[['YearsCodePro', 'ExperienceLevel']].head()
```
